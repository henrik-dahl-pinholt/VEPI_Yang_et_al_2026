# MS2 correlation fit plots

This notebook uses `1_1_plot_MS2_corrfit.py` to load the cached fit output and render the fitted autocovariance functions and mean intensities.

```python
from pathlib import Path
import importlib.util
import sys

script_path = Path("1_1_plot_MS2_corrfit.py").resolve()
spec = importlib.util.spec_from_file_location("plot_ms2_corrfit", script_path)
plot_ms2_corrfit = importlib.util.module_from_spec(spec)
sys.modules[spec.name] = plot_ms2_corrfit
spec.loader.exec_module(plot_ms2_corrfit)

bundle = plot_ms2_corrfit.load_fit_bundle()
bundle.summary_path, bundle.arrays_path
```

```python
fig_acf = plot_ms2_corrfit.plot_acf_fits(bundle)
fig_acf
```

```python
fig_means = plot_ms2_corrfit.plot_means(bundle)
fig_means
```

```python
plot_ms2_corrfit.save_figures(bundle, plot_ms2_corrfit.DEFAULT_OUTPUT_DIR)
```
